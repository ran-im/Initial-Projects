```python
from selenium import webdriver

# Specify the path to the Microsoft Edge WebDriver executable
edge_path = 'C:/Users/imran/Downloads/edgedriver_win64/msedgedriver.exe'

# Create a Service instance with the specified path
edge_service = webdriver.EdgeService(executable_path=edge_path)

# Instantiate the Edge driver with the EdgeService instance
driver = webdriver.Edge(service=edge_service)

# Navigate to a website
driver.get('https://in.indeed.com/jobs?q=Business+Analyst&l=Delhi&from=searchOnHP&vjk=9e6bd3327dfb230f')
```

```python
html_content = driver.page_source
```

```python
print(html_content)
```

```python
import requests 
from bs4 import BeautifulSoup
import html5lib
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup.prettify)
```

```python
title = soup.title
print(title)
```

```python
para = soup.find_all('p')
print(para)
```

```python
anchors = soup.find_all('a')
all_links = set()
for link in anchors:
    if(link.get('href') != '#'):
        linkText = "https://in.indeed.com" + link.get('href')
        all_links.add(link)
    print(linkText)
```
